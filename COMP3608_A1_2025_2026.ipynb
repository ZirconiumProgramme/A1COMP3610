#Question 1

The Scroll of Electrical Power of the Naka River Dam!

Observe! Behold the ancient scroll, etched with the secrets of power consuption at different levels of the Naka River Dam's output!



|Level|Power |
|---|---|
| 2 | 1 |
| 3 | 3 |
| 5 | 3 |
| 7 | 5 |


Our Hokage, Tsunade-sama herself, along with Orochimaru (yes, even he has a vested interest in this!), seeks to unlock the hidden formula

\begin{equation*}
Power = β_0 + β_1*Level
\end{equation*}

They want to fit this jutsu to the data in a way that minimizes the sum of the absolute deviations between the predicted and observed power requirements.

(a) Write a mathematical program to solve the above problem. **[8 marks]**
\
(b) Write a program using pulp to solve (a). **[12 marks]**

Answer for a) We must minimize the sum of the absolute derivations between the predicted and observed power requirements so. |Pe - Po| ≤ e since the error must be equal to or greater than the difference between the expected and observed also the error must be non negative so e > 0. given P = B0 + B1 * L, where B0, B1 ∈ R

Recall the definition of the absolute function

|x| = x when x ≥ 0 and |x| = -(x) when x < 0.

So we are left with the cosntraints

Pe - Po ≤ e  and Po - Pe ≤ e
using P = B0 + B1 * L we get finally

P - (B0 + B1 * L) ≤ e and
B0 + B1 * L - P ≤ e

Since we must minimize all of the errors we are left can construct the objective function:

Min∑Ei where error, e = |Poi - Pe| and i = [1, 2, 3, 4]

Objective Function

Min∑Ei, Ei > 0

Constraints

Pi - (B0 + B1 * Li) ≤ e and
B0 + B1 * Li - Pi ≤ e

where i = [1, 2, 3, 4]


```python
import pulp as p

levels = [2, 3, 5, 7]
powers = [1, 3, 3, 5]

prob = p.LpProblem("Assignment_1_Q1", p.LpMinimize)


B = p.LpVariable("B", [0, 1], lowBound= None)
A = p.LpVariable("E", range(len(powers), lowBound = 0))

prob += p.lpSum(A[i] for i in range(len(powers)))

for i in range(len(powers)):
  prob += powers[i] - (B[0] + B[1] * levels[i]) <= A[i]
  prob += -(powers[i] - (B[0] + B[1] * levels[i])) <= A[i]

  prob.solve()
  print("Status:", p.LpStatus[prob.status])
  print("B0:", B[0].varValue)
  print("B1:", B[1].varValue)
```

#Question 2

Solve Question 1 using gradient descent. Your solution should be within 0.2 of its optimal $β_0$ and $β_1$. **[20 marks]**

```python
import torch as th

levels = th.tensor([2, 3, 5, 7])
powers = th.tensor([1, 3, 3, 5])

B0 = th.rand(1, requires_grad = True)
B1 = th.rand(1, requires_grad = True)

optimizer = th.optim.SGD([B0, B1], lr = 0.001)

for epoch in range(20000):
  iterations = B0 + B1 * levels
  loss = th.sum(th.abs(powers - iterations))
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

print("B0:", B0.item())
print("B1:", B1.item())

```
